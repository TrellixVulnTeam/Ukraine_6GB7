```python
def getLinks(df):
    pattern = re.compile(r"(?<=://)(.*)(?=</url)")
    df['Links'] = df['message'].str.extract(pattern, expand=False)
```

```python
def extract_emoji(df):
    df["countries"] = ""
    for index, row in df.iterrows():
        for emoji in EMOJIS:
            if emoji in row["message"]:
                row["message"] = row["message"].replace(emoji, "")
                row["countries"] += emoji

edu_info = extract_emoji(df)
print(df.to_string())
```
